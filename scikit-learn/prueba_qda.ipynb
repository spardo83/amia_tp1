```python
# QDA sobre Iris en scikit-learn: notebook paso a paso

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1) Cargar dataset como DataFrame

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

df = pd.DataFrame(X, columns=feature_names)
df["target"] = y
df["target_name"] = df["target"].map({i: n for i, n in enumerate(target_names)})

print("Primeras filas del dataset Iris:")
print(df.head())
```

```python
# 2) Train / Test split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# 3) Entrenar QDA como clasificador
#    (si tenés problemas de matrices singulares en otros datasets, probá reg_param>0)

qda = QuadraticDiscriminantAnalysis(reg_param=0.0)
qda.fit(X_train, y_train)
```

```python
# 4) Evaluación básica

y_pred = qda.predict(X_test)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"\nAccuracy en test: {acc:.4f}\n")
print("Reporte de clasificación:\n")
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# 5) Validación cruzada estratificada (5 folds) en todo el dataset

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(QuadraticDiscriminantAnalysis(), X, y, cv=cv)
print("CV 5-fold scores:", cv_scores)
print("CV mean:", cv_scores.mean(), "±", cv_scores.std())

# 6) Visualización: regiones de decisión en 2D usando dos features
#    Usamos 'petal length (cm)' y 'petal width (cm)' por ser las más separables.
feat_idx = (2, 3)  # columnas 2 y 3
X2 = X[:, feat_idx]
feat_names = [feature_names[i] for i in feat_idx]
```

```python
# 6) Visualización: regiones de decisión en 2D usando dos features
#    Usamos 'petal length (cm)' y 'petal width (cm)' por ser las más separables.

feat_idx = (2, 3)  # columnas 2 y 3
X2 = X[:, feat_idx]
feat_names = [feature_names[i] for i in feat_idx]

# Entrenar QDA solo con estas 2 features para graficar regiones
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y, test_size=0.3, random_state=42, stratify=y
)
qda_2d = QuadraticDiscriminantAnalysis()
qda_2d.fit(X2_train, y2_train)

# Malla para regiones de decisión
x_min, x_max = X2[:, 0].min() - 0.5, X2[:, 0].max() + 0.5
y_min, y_max = X2[:, 1].min() - 0.5, X2[:, 1].max() + 0.5
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)
Z = qda_2d.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(6, 5))
# Fondo con regiones
plt.contourf(xx, yy, Z, alpha=0.2)
# Puntos reales
for i, name in enumerate(target_names):
    mask = y == i
    plt.scatter(X2[mask, 0], X2[mask, 1], label=name, edgecolor="k", s=40)
plt.xlabel(feat_names[0])
plt.ylabel(feat_names[1])
plt.title("QDA - Regiones de decisión (2 features)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

```python
# 7) Matriz de confusión (test del modelo con las 4 features)

plt.figure(figsize=(4, 4))
plt.imshow(cm)
plt.title("Matriz de confusión (Test)")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.xticks(ticks=[0,1,2], labels=target_names, rotation=45, ha="right")
plt.yticks(ticks=[0,1,2], labels=target_names)
for (i, j), v in np.ndenumerate(cm):
    plt.text(j, i, str(v), ha='center', va='center')
plt.tight_layout()
plt.show()
```
